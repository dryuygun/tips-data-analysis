```
# Not hesaplama fonksiyonu
def not_hesaplama():
  def not_değeri(ders_adi):
    while True:
      try:
        not_degeri = float(input(f"{ders_adi} Notu (0-100): "))
        if 0 <= not_degeri <= 100:
          return not_degeri
        else:
          print("⚠️ Lütfen 0 ile 100 arasında bir değer giriniz.")
      except ValueError:
          print("Lütfen geçerli bir not giriniz!")

  sayi1 = not_değeri("Türkçe Notu: ")
  sayi2 = not_değeri("Matematik Notu: ")
  sayi3 = not_değeri("Fizik Notu: ")
  sayi4 = not_değeri("Kimya Notu: ")
  sayi5 = not_değeri("Biyoloji Notu: ")

  ortalama = (sayi1 + sayi2 + sayi3 + sayi4 + sayi5) / 5

  if ortalama >= 50:
    print(f"Sonuç: {ortalama} Notuyla Geçtiniz")
  else:
    print(f"Sonuç: {ortalama} Notuyla Kaldınız")


# Fonksiyonu çağır
not_hesaplama()
```

# Mini Veri Analizi Raporu

•⁠  ⁠Veride toplam *{df.shape[0]} gözlem, *{df.shape[1]} değişken** var.
•⁠  ⁠Ortalama bahşiş miktarı *{round(df['tip'].mean(), 2)}* dolar.
•⁠  ⁠En yüksek bahşiş *{df['tip'].max()}* dolar.
•⁠  ⁠Görselleştirme sonuçlarına göre:
  - Çoğu bahşiş *2 ile 4 dolar arasında* veriliyor.
  - Toplam hesap arttıkça bahşiş de genellikle artıyor.
  - Kadın ve erkek müşterilerin bahşiş davranışlarında belirgin farklar gözlenmiyor (detaylı test gerekebilir).

```
# Müşteri Segmentasyonu

# pandas ve seaborn'dan yerleşik bir veri seti alalım
import matplotlib.pyplot as plt
import seaborn as sns

# 'tips' veri setini yükle (restoran bahşiş verisi)
df = sns.load_dataset("tips")

# İlk 5 satırı görüntüle
df.head()

#Segment belirleyen fonksiyon

def segment_belirle(tip):
  if tip < 2:
    return "Düşük"
  elif 2 <= tip < 4:
    return "Orta"
  else:
    return "Yüksek"

# Yeni sütun ekle
df['segment'] = df['tip'].apply(segment_belirle)

# İlk birkaç satırı görelim
df[['tip', 'segment']].head()


# ANALİZ - Her segmentte kaç müşteri var?
df['segment'].value_counts()

# ANALİZ - Her segmentin ortalama bahşiş ve hesap bilgisi
df.groupby('segment')[['total_bill', 'tip']].mean()

# GÖRSELLEŞTİRME
sns.boxplot(data=df, x='segment', y='tip', palette='pastel')
plt.title("Segmentlere Göre Bahşiş Dağılımı")
plt.show()
```

```
# ÖDEV - GÖRSELLEŞTİRME
# ÖDEV : Sigara için ve içmeyen müşteriler arasında fark var mı?

# pandas ve seaborn'dan yerleşik bir veri seti alalım
import pandas as pd
import seaborn as sns

# 'tips' veri setini yükle (restoran bahşiş verisi)
df = sns.load_dataset("tips")

# İlk 5 satırı görüntüle
df.head()

def segment_belirle(tip):
  if tip < 2:
    return "Düşük"
  elif 2 <= tip < 4:
    return "Orta"
  else:
    return "Yüksek"

# Yeni sütun ekle
df['segment'] = df['tip'].apply(segment_belirle)

pd.crosstab(df['segment'], df['smoker'])
df.groupby(['segment', 'smoker'])[['tip']].mean()

sns.barplot(data=df, x='segment', y='tip', hue='smoker', palette='pastel')
plt.title("Segmentlere Göre Bahşiş Dağılımı")
plt.xlabel("Müşteri Segmenti")
plt.xlabel("Ortalama Bahşiş")
plt.show()
```

### Segmentlere Göre Sigara Durumunun Bahşiş Üzerindeki Etkisi

•⁠  ⁠Düşük segmentte sigara içip içmeme durumu bahşişi pek etkilemiyor.

•⁠  ⁠Orta segmentte içmeyenler hafifçe daha fazla bahşiş bırakıyor.

•⁠  ⁠Yüksek segmentte de içmeyenler bir adım önde görünüyor, ama fark küçük.

•⁠  ⁠Genel olarak sigara içme durumu bahşiş davranışını çok ciddi şekilde değiştirmiyor gibi görünüyor.
